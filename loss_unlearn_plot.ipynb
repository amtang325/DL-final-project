```python
import numpy as np
import matplotlib.pyplot as plt

cycle = [0,1,2,3,4]

train_loss_forget = [np.nan, 161.7005, np.nan, 112.5944, np.nan]
train_acc_forget = [np.nan, 0.0943, np.nan, 0.0415, np.nan]
train_loss_retain = [np.nan, np.nan, 2.9922, np.nan, 4.9238]
train_acc_retain = [np.nan, np.nan, 0.6970, np.nan, 0.6761]
val_loss_forget = [1.3367, 112.6714, 1.9419, 147.7588, 2.6235]
val_acc_forget = [0.5345, 0.0690, 0.4397, 0.0690, 0.4310]
val_loss_retain = [0.9046, 29.1405, 0.8538, 37.0493, 0.9635]
val_acc_retain = [0.6490, 0.4204, 0.6883, 0.3879, 0.6604]
```

```python
plt.figure()
plt.plot(cycle, train_loss_forget, marker='o', label='Forget')
plt.plot(cycle, train_loss_retain, marker='s', label='Retain')
plt.title('Training Loss')
plt.xlabel('Unlearning Stage')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(cycle, train_acc_forget, marker='o', label='Forget')
plt.plot(cycle, train_acc_retain, marker='s', label='Retain')
plt.title('Training Accuracy')
plt.xlabel('Unlearning Stage')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(cycle, val_loss_forget, marker='o', label='Forget')
plt.plot(cycle, val_loss_retain, marker='s', label='Retain')
plt.title('Validation Loss')
plt.xlabel('Unlearning Stage')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(cycle, val_acc_forget, marker='o', label='Forget')
plt.plot(cycle, val_acc_retain, marker='s', label='Retain')
plt.title('Validation Accuracy')
plt.xlabel('Unlearning Stage')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
loss_diff = [x - y for x, y in zip(val_loss_forget, val_loss_retain)]
acc_diff  = [x - y for x, y in zip(val_acc_retain, val_acc_forget)]

plt.figure()
plt.plot(cycle, loss_diff, marker='o')
plt.title('Validation Loss Difference (Forget – Retain)')
plt.xlabel('Unlearning Stage')
plt.ylabel('Loss Difference')
plt.show()

plt.figure()
plt.plot(cycle, acc_diff, marker='o')
plt.title('Validation Accuracy Difference (Retain – Forget)')
plt.xlabel('Unlearning Stage')
plt.ylabel('Accuracy Difference')
plt.show()
```
